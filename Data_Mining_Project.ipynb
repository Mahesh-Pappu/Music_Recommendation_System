```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score



# Selecting relevant columns
X = df[['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']]

# Initialize lists to store inertia and silhouette scores
inertia = []
silhouette_scores = []

# Range of k values to try
k_range = range(2, 11)

# Iterate through each k value
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)

    # Append inertia and silhouette score to the lists
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plotting the elbow curve
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')

# Plotting silhouette scores
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Selecting relevant columns
X = df[['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize Hierarchical clustering with the desired number of clusters
n_clusters = 6  # You can change this to the desired number of clusters
hierarchical_clustering = AgglomerativeClustering(n_clusters=n_clusters)

# Fit the hierarchical clustering model to the data
hierarchical_clustering.fit(X_scaled)

# Create linkage matrix using complete linkage
Z = linkage(X_scaled, method='complete')

# Plotting the dendrogram
plt.figure(figsize=(12, 6))
plt.title('Hierarchical Clustering Dendrogram')
dendrogram(Z, truncate_mode='level', p=3)
plt.xlabel('Number of points in node (or index of point if no parenthesis)')
plt.ylabel('Distance')
plt.show()
```

```python
# Assuming hierarchical_clustering.labels_ contains the cluster labels for each data point
cluster_labels = hierarchical_clustering.labels_

# Create a DataFrame with the cluster labels and original features
cluster_data = pd.DataFrame(X, columns=['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%'])
cluster_data['Cluster'] = cluster_labels

# Group by cluster and calculate the mean values of features
cluster_means = cluster_data.groupby('Cluster').mean()

print(cluster_means)
```
